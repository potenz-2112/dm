```python
import pandas as pd
import numpy as np
import random
```

```python
missing_values=["NaN", "NA", "<null>"]
df=pd.read_csv("pima_Missing_values.csv", na_values=missing_values)
```

```python
df
```

```python
df.isnull()
```

```python
# count of missing values in each column.

count = df.isnull().sum()
count
```

```python
# percentage of missing values in each column.

count/len(df) * 100
```

```python
# removing all columns with more than 10.5% missing values.
# threshold (thresh) takes value of number of non-NA values. So we require 89.5%
# non-NA values. Hence thresh=89.5%.
# So 2 columns will be dropped, i.e. bloodpressure & diabetespedigree.

limitper=0.895*len(df)
df1=df.dropna(axis=1, thresh=limitper)
df1
```

```python
# removing all rows having missing values.

df2=df.dropna(axis=0)
df2
```

```python
# adding random values in each column between their range.
df3=df.copy()

for column in df3.columns[:-2]:
    df3[column]=df3[column].fillna(random.uniform(df3[column].min(), df3[column].max()))
    
df3
```

```python
# filling each missing value in column with its mean value.

df4 = df.copy()

for column in df4.columns[:-2]:
    df4[column]=df4[column].fillna(df4[column].mean())
    
df4
```

```python
# Interpolation is the process of finding a value between two points 
# on a line or curve. It is mostly used to predict the unknown values 
# for any geographical related data points.

df5 = df.copy()

df5=df.interpolate(method="linear")
df5
```
