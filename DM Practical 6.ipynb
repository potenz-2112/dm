```python
import numpy as np
import pandas as pd
import random


def sigmoid(x):
    return 1.0/(1.0 + np.exp(-x))

def sigmoid_der(x):
    return x*(1.0 - x)


class NN:
    def __init__(self, inputs):
        self.inputs = inputs
        self.l=len(self.inputs)#4 inputs
        self.li=len(self.inputs[0])# 2 values 0 and 1
        
        #generating weights
        
        self.w1=np.random.random((self.li, self.l))
        self.w2=np.random.random((self.l, 1))

    def think(self, inp):
        s1=sigmoid(np.dot(inp, self.w1))
        s2=sigmoid(np.dot(s1, self.w2))
        return s2
    
    
    def train(self, inputs,outputs, ite):
        for i in range(ite):
            l0=inputs
            l1=sigmoid(np.dot(l0, self.w1))
            l2=sigmoid(np.dot(l1, self.w2))
            
            #back propagation
            
            l2_err=outputs - l2
            l2_delta = np.multiply(l2_err, sigmoid_der(l2))

            l1_err=np.dot(l2_delta, self.w2.T)
            l1_delta=np.multiply(l1_err, sigmoid_der(l1))
            #updating weights
            self.w2+=np.dot(l1.T, l2_delta)
            self.w1+=np.dot(l0.T, l1_delta)

inputs=np.array([ [0,0], [0,1], [1,0], [1,1] ])
outputs=np.array([ [0],[1],[1],[0] ])
n=NN(inputs)
print("Before training...")
print(n.think(inputs))
n.train(inputs, outputs, 10000)
print("After training...")
print(n.think(inputs))
```
