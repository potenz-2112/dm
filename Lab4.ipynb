```python
import pandas as pd
import numpy as np
df=pd.read_csv('itemset.csv')
df
```

```python
#data seperation
trans_df=df.item_list.str.split(',')
trans_df
```

```python
#listing of all the items in the transactions
init=[]
for i in trans_df:
  for q in i:
    if(q not in init):
      init.append(q)

init=sorted(init)
print(init)
    
length = len(init)
print("length:",length)
```

```python
#support count defining
sp = 0.4
s = int(sp*len(init))
s
```

```python
from pandas.core.arrays import string_
from collections import Counter


# initializing the counter for counting the number of times the items are present in the transactions
c = Counter()
for i in init:
    for d in trans_df:
        if(i in d):
            c[i]+=1
print("C1:")
for i in c:
    print(str([i])+": "+str(c[i]))
print()

#Counter for ruling out the entries which are less than the support count
l = Counter()
for i in c:
    if(c[i] >= s):
        l[frozenset([i])]+=c[i]
print("L1:")
for i in l:
    print(str(list(i))+": "+str(l[i]))
print()
pl = l
pos = 1

#defining the subset function to find the subsets of the given set
import itertools
def findsubsets(s, n):
    return list(itertools.combinations(s, n))

# for item numbers greater than 2
# finding their subsets using the function defined above
for count in range (2,length):
    nc = set()
    temp_list = list(l)
    
    for i in range(0,len(temp_list)):
        
        for j in range(i+1,len(temp_list)):
            
            t = temp_list[i].union(temp_list[j])
            
            if(len(t) == count):
                nc.add(temp_list[i].union(temp_list[j]))
    
    nc = list(nc)
    c = Counter()
    
    for i in nc:
       
       arr=findsubsets(list(i),count-1)
       print(arr)
       cr=0

       for j in arr:
            je=set(j)
            result=temp_list.count(je)
            print(result)
       
            if result>0:
                cr+=1
            else:
                cr-=1
        
            if(cr>=count):
                c[i] = 0
                for q in trans_df:
                    tem = set(q)
                    
                    if(i.issubset(tem)):
                        c[i]+=1
            else:
                x=list(nc)
                x.remove(i)

    print("C"+str(count)+":")
    
    for i in c:
        
        print(str(list(i))+": "+str(c[i]))
    
    print()

    #finding and deleting itemsets which are less than the support count
    l = Counter()
    
    for i in c:
        
        if(c[i] >= s):
        
            l[i]+=c[i]
    
    print("L"+str(count)+":")
    
    for i in l:
        
        print(str(list(i))+": "+str(l[i]))
    
    print()
    
    if(len(l) == 0):
    
        break
    
    pl = l
    pos = count
```

```python
print("Result: ")
print("L"+str(pos)+":")
for i in pl:
    print(str(list(i))+": "+str(pl[i]))
print()
```
